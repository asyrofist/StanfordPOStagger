```python
def lenCheck(split_train_2, split_test_2):
    sum = 0
    for i in range(len(split_train_2)):
        if(len(split_train_2[i].split(" ")) != len(split_test_2[i].split(" "))):
            print(str(i + 1) + " - " + str(len(split_train_2[i].split(" ")) - len(split_test_2[i].split(" "))))
            sum = sum + abs(len(split_train_2[i].split(" ")) - len(split_test_2[i].split(" ")))
    print(sum)
```
